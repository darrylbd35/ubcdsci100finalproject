```r
library(tidyverse)
```

```r
sessions <- read_csv("data/sessions.csv")
players <- read_csv("data/players.csv")
```

```r
sessions_and_players <- inner_join(sessions, players)
sessions_and_players
```

```r
average_hr_played <- sessions_and_players |> 
  summarize(average_hr_played = round(mean(played_hours, na.rm = TRUE), 2))
average_hr_played
```

```r
average_gender <- sessions_and_players |>
  group_by(gender) |>
  summarize(count = n()) |>
  mutate(percentage = round(count / sum(count) * 100, 2))
average_gender
```

```r
average_experience <- sessions_and_players |> 
  group_by(experience) |>
  summarize(count = n()) |>
  mutate(percentage = round(count / sum(count) * 100, 2)) |>
  arrange(desc(count))
average_experience
```

```r
average_played_hours <- sessions_and_players |> 
  summarize(average_played_hours = round(mean(played_hours, na.rm = TRUE), 2))
average_played_hours
```

```r
average_age <- sessions_and_players |>
    summarize(average_age = round(mean(Age, na.rm = TRUE), 2))
average_age
```

```r
sessions <- sessions |>
    mutate(original_end_time = as.numeric(as.POSIXct(end_time, format = "%d/%m/%Y %H:%M")),
           original_start_time = as.numeric(as.POSIXct(start_time, format = "%d/%m/%Y %H:%M")))
session_duration <- sessions |>
    mutate(duration = original_end_time - original_start_time) |>
    select(duration)
sessions_summary <- session_duration |>
    summarize(min_duration = round(min(duration, na.rm = TRUE), 2),
              Q1_duration = round(quantile(duration, 0.25, na.rm = TRUE), 2),
              median_duration = round(median(duration, na.rm = TRUE), 2),
              Q2_duration = round(quantile(duration, 0.75, na.rm = TRUE), 2),
              max_duration = round(max(duration, na.rm = TRUE), 2),
              mean_duration = round(mean(duration, na.rm = TRUE), 2))
sessions_summary
```
